```python
import nltk
#nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
sid = SentimentIntensityAnalyzer()
#df = pd.read_csv('Data/GD_INS_df_wlabels.csv')
#df = pd.read_csv('1collapsed_df_all_companies.csv')
df = pd.read_pickle("./collapsed_review_data.pkl")
```

```python
sid.polarity_scores(df.loc[0]['cons'])
```

```python
#pd.set_option("display.max_colwidth",1000)
#df.head(10)
```

```python
df['con_scores'] = df['cons'].apply(lambda con_scores: sid.polarity_scores(con_scores))
df['cons_compound']  = df['con_scores'].apply(lambda score_dict: score_dict['compound'])
df['cons_negative']  = df['con_scores'].apply(lambda score_dict: score_dict['neg'])
df['cons_positive']  = df['con_scores'].apply(lambda score_dict: score_dict['pos'])
df['pro_scores'] = df['pros'].apply(lambda pro_scores: sid.polarity_scores(pro_scores))
df['pros_compound']  = df['pro_scores'].apply(lambda score_dict: score_dict['compound'])
df['pros_negative']  = df['pro_scores'].apply(lambda score_dict: score_dict['neg'])
df['pros_positive']  = df['pro_scores'].apply(lambda score_dict: score_dict['pos'])

#to deal w NAs
def score_polarity(con_score):
    if con_score == np.nan:
          return np.nan
    else:
          return sid.polarity_scores(con_score)
        
df['con_scores'] = df['cons'].apply(score_polarity)
```

```python
np.mean(df['cons_negative'])
```

```python
np.mean(df['pros_compound'])
```
